```python
dataset1 = pd.read_csv("Wine.csv", index_col=None)

# Create a copy of the dataset
df2 = dataset1.copy()

# Convert categorical variables to dummy/indicator variables
df2 = pd.get_dummies(df2, dtype=int ,drop_first=True)

# Separate the independent and dependent variables
indep_X = df2.drop('Customer_Segment', axis=1)  # Drop the dependent variable column
dep_Y = df2['Customer_Segment']                  # Set 'Customer_Segment' as the dependent variable
```
